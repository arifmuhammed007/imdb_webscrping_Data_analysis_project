```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import pandas as pd

# Start Chrome
driver = webdriver.Chrome()
driver.get("https://www.imdb.com/chart/top/")
time.sleep(3)

movies = []

# Find all movie list items
rows = driver.find_elements(By.CSS_SELECTOR, "li.ipc-metadata-list-summary-item")

for row in rows:
    # Title (clean: remove "1. " part)
    raw_title = row.find_element(By.CSS_SELECTOR, "h3").text
    title = raw_title.split(". ", 1)[-1]
    
    # All metadata items
    metadata_items = row.find_elements(By.CSS_SELECTOR, "span.cli-title-metadata-item")
    
    # Initialize variables to handle missing data
    year = ""
    duration = ""
    content_rating = ""

    # Check the number of metadata items
    if len(metadata_items) >= 1:
        year = metadata_items[0].text
    if len(metadata_items) >= 2:
        duration = metadata_items[1].text
    if len(metadata_items) >= 3:
        content_rating = metadata_items[2].text
    
    # Rating
    try:
        rating = row.find_element(By.CSS_SELECTOR, "span.ipc-rating-star--rating").text
    except:
        rating = ""
    
    # Votes
    try:
        votes = row.find_element(By.CSS_SELECTOR, "span.ipc-rating-star--voteCount").text
    except:
        votes = ""
    
    movies.append([title, year, duration, content_rating, rating, votes])

driver.quit()

# Convert to DataFrame
df = pd.DataFrame(movies, columns=["Title", "Year", "Duration", "Content Rating", "Rating", "Votes"])

```
